```python
import pandas as pd
import matplotlib.pyplot as plt

# Count songs per genre
genre_counts = df['genre'].value_counts()

plt.figure(figsize=(10,16))
genre_counts.plot(kind='barh')

plt.title("Genre Distribution")
plt.xlabel("Count")
plt.ylabel("Genre")

plt.show()
```
